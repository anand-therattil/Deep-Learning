```python
import os
import sys
import librosa
import matplotlib.pyplot as plt
import numpy as np

def feature_extraction(audio_path):
    x, sr = librosa.load(audio_path)
    print(x.shape)
    mfccs = librosa.feature.mfcc(y=x, n_mfcc=13, sr=sr)
    D = librosa.feature.delta(mfccs)
    DD = librosa.feature.delta(mfccs, order=2)
    return mfccs,D,DD,x,sr,

def plot_spectrograms(input_array,sr):
  plt.figure(figsize=(25,5))
  librosa.display.specshow(input_array,x_axis="time",sr=sr)
  plt.colorbar(format="%+2f")
  plt.show()

audio_path = r"/content/f01.wav"
mfccs,D,DD,x,sr=feature_extraction(audio_path)
feature = np.concatenate([mfccs,D,DD])
plot_spectrograms(mfccs,sr)

```
